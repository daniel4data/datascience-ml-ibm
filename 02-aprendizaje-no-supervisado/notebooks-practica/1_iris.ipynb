```python
### Primer código del pdf 2 - flores iris
from sklearn.preprocessing import MinMaxScaler
from sklearn import datasets # 'datasets' contiene funciones para cargar conjuntos de datos estándar (para practicar)
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris() # cargamos el dataset de iris
```

```python
X = np.array(iris.data) # convierte los datos que están en iris.data en un objeto tipo ndarray de NumPy, es decir, una matriz de datos numéricos
# 'X' contiene los datos de cada una de las flores, sin su especie. Es decir, contiene las características de cada flor
y = np.array(iris.target) # 'Y' contiene a las especies (clases)
```

```python
# Ploteamos las últimas 2 columnas
# X = array (matriz) de [150, 4]
plt.scatter(X[:,2], # todas las filas, 3er columna
            X[:,3], # todas las filas, 4ta columna
            c=y) # 
plt.xlabel('petal lenght')
plt.ylabel('petal width')
plt.show()
```

```python
# Nota que la variable petallength tiene valores que van desde 1 hasta 7, mientras que en petalwidth van de 0 a 2.5.
```

```python
# necesitamos escalar, debido a que hay columnas que tienen magnitudes que varían mucho (el tamaño, por ejemplo)
# aplicamos MinMaxScaler
scaler = MinMaxScaler()
X2 = scaler.fit_transform(X) # eso transforma todos los valores para que estén entre 10 y 20, por si lo necesitas (ya no de 0 a 1), aunque esto es indistinto
```

```python
plt.scatter(X2[:,2], # todas las filas, 3er columna
            X2[:,3], # todas las filas, 4ta columna
            c=y) # 
plt.xlabel('petal lenght')
plt.ylabel('petal width')
plt.show()
```
