```python
### 4. Ejemplo de agrupamiento usando el algoritmo K-Means ###
from sklearn import datasets
import numpy as np
```

```python
# 1) Cargando iris
iris = datasets.load_iris()
```

```python
# Cargando los datos sin la clase
X = np.array(iris.data) # Es decir, aún no cargamos "target"; de lo contrario, el algoritmo identificará las clases verdaderas
```

```python
# 2) Definir cuántos grupos queremos
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3) # Le estamos indicando que queremos formar 3 grupos
```

```python
# 3) Entrenamos el modelo:
kmeans.fit(X) # Aquí se realiza todo el proceso: elige centroides, asigna puntos, recalcula centroides y repite hasta converger
```

```python
# 4) Obtenemos los grupos a los que cada punto fue asignado
labels = kmeans.predict(X) # 'predict()' no es una predicción clásica, sino que indica a qué grupo (0, 1 y 2) pertenece cada flor según kmeans
```

```python
# 5) Visualizamos
import matplotlib.pyplot as plt
plt.scatter(X[:,2], X[:,3], c=labels)
# X[:, 2] selecciona la tercera columna (índice 2) → petal length.
# X[:, 3] selecciona la cuarta columna (índice 3) → petal width.
# c=labels colorea cada punto según su grupo asignado por K-means (labels es un array de 150

plt.xlabel('Petal_length')
plt.ylabel('Petal_width')
# Añade etiquetas a los ejes para interpretar el gráfico

plt.show()
```

```python
# 6) Comparar con las clases verdaderas
y = np.array(iris.target) # Ahora sí cargamos las 3 clases reales (0=setosa, 1=versicolor, 2=virginica)
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(9, 4)) # Crea dos subgráficos en la misma figura, colocados horizontalmente (ncols=2)
# ax0, ax1 son los ejes individuales donde se va a graficar.

# Subgráfico 1: Etiquetas reales
ax0.scatter(X[:, 2], X[:, 3], c=y)
ax0.set_title('Clases originales')
ax0.set_xlabel(iris.feature_names[2]) # Usa los nombres originales de las variables del dataset (iris.feature_names) para etiquetar los ejes
ax0.set_ylabel(iris.feature_names[3])

# Subgráfico 2 – Grupos de K-means
ax1.scatter(X[:, 2], X[:, 3], c=labels)
ax1.set_title('Grupos obtenidos')
ax1.set_xlabel(iris.feature_names[2])
ax1.set_ylabel(iris.feature_names[3])

plt.show()
```
