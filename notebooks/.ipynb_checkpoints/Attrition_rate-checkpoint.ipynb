```python
# ================ Reto 2: Attrition Rate ================ #
'''En este reto, tu tarea es agrupar a los empleados y brindar una interpretación de los grupos
obtenidos, tomando como una de las var iables más relevantes para la descripción de los
resultados la ratio de deserción (Attrition_rate).'''
```

```python
'''Antes de aplicar K-Means, necesitamos preparar los datos con estos pasos:

1. Eliminar columnas irrelevantes: como identificadores únicos o variables sin variación (por ejemplo, si todos tienen el mismo valor).
2. Crear nuevas variables útiles: por ejemplo, calcular cuántos años lleva un empleado en la empresa.
3. Convertir variables categóricas a numéricas: ya que K-Means solo trabaja con números. Esto se hace con codificadores como:
    * OrdinalEncoder si hay un orden (por ejemplo: educación).
    * OneHotEncoder si no hay orden (por ejemplo: departamento).
4. Escalar las variables numéricas: para que todas estén en un mismo rango y una no domine a las otras (por ejemplo, el sueldo puede tener valores mucho mayores que otras variables).'''
```

```python
# ==== 1. SETUP INICIAL ==== #
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('data/Train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Identificamos 'Attrition_rate' que es nuestra variable de salida, la cual es tipo float
# Las que son categóricas -> codificar (OneHotEncoder u OrdinalEncoder)
# Las que son numérticas -> (float e int) -> normalizarlas
# Hay que analizar las variables 'VAR' -> Analizar su correlación con la variable dependiente
```

```python
df.columns
```

```python
# ==== 2. LIMPIEZA INICIAL DEL DATASET ==== #
# Eliminamos columnas irrelevantes -> Aquellas que no explican patrones, son constantes (invariantes) y 
df = df.drop(['Employee_ID'], axis=1)
```

```python
# Revisamos valores NA
df.isnull().sum()
```

```python
# == Age == #
# Dado que 'Age' no es una variable continua, desistimos de usar la media. Optamos por la mediana, pero para ello hay que revisar si la distribución es normal o sesgada
df['Age'].hist(bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.show()
```

```python
# Vemos que los valores de 'Age' no se distribuyen normal -> Empleamos la mediana para rellenar valores NA
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
```

```python
# == Work_Life_balance == #
# En el caso de 'Work_Life_balance', parece representar una escala ordinal discreta, de 1-5. En caso de ser así, conviene usar la moda
df['Work_Life_balance'].value_counts()
```

```python
sns.boxplot(x='Work_Life_balance', y='Attrition_rate', data=df)
plt.title('Attrition rate por nivel de Work-Life Balance')
plt.show()
```

```python
# Dado que los nulos son pocos, y la variable aporta aunque sea ligeramente, conviene imputar
wlb_mode = df['Work_Life_balance'].mode().iloc[0]
df['Work_Life_balance'] = df['Work_Life_balance'].fillna(wlb_mode)
```

```python
# == VAR2/VAR4 == #
df[['Attrition_rate','VAR2','VAR4']].corr()
```

```python
df[['Attrition_rate', 'VAR2', 'VAR4']].dropna().corr()
```

```python
# Dado que la correlación es baja, optamos por eliminarlas
df = df.drop(['VAR2', 'VAR4'],axis=1)
```

```python
# == Time_of_service == #
df['Time_of_service'].hist(bins=30)
plt.xlabel('Time of service')
plt.ylabel('Frequency')
plt.show()

sns.boxplot(x=df['Time_of_service'])
plt.show()
```

```python
median_tos = df['Time_of_service'].median()
df['Time_of_service'] = df['Time_of_service'].fillna(median_tos)
```

```python
# == Pay_Scale == #
median_payscale = df['Pay_Scale'].median()

df['Pay_Scale'] = df['Pay_Scale'].fillna(median_payscale) # # Rellenar los NaN con esa mediana
```

```python
# ==== 3. TRANSFORMACIÓN DE VARIABLES CATEGÓRICAS ==== #
# Convertimos las variables tipo 'object' en numéricas

# ==== OneHotEncoder ==== # -> Categorías SIN ORDEN
df.select_dtypes(include='object')
```

```python
# Dado que no tienen orden lógico, empleamos OneHotEncoder
df_model = df.copy()
df_model = pd.get_dummies(df, columns=[
    'Gender',
    'Relationship_Status', 
    'Hometown', 
    'Unit', 
    'Decision_skill_possess', 
    'Compensation_and_Benefits'
], drop_first=True) # omitimos la primera categoría de cada variable
```

```python
# Convertimos booleanos a enteros
df_model = df_model.astype({col:int for col in df_model.columns if df_model[col].dtype == 'bool'})
```

```python
# Guardamos Attrition_rate antes de escalar
attrition = df_model['Attrition_rate']
```

```python
# ==== 4. ESCALAR VARIABLES NUMÉRICAS ==== #
# == StandardScaler == # -> Lo recomendado para K-Means

# Seleccionamos todas las variables numéricas
df.select_dtypes(include=[np.number])
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns.drop('Attrition_rate') # Excluimos 'Attrition_rate'
```

```python
scaler = StandardScaler()
df_scaled = df_model.copy()
df_scaled[num_cols] = scaler.fit_transform(df_model[num_cols]) # # Aplicamos el escalado sólo a las columnas numéricas seleccionadas
```

```python
df_scaled[num_cols].describe().round(2) # Verificamos el escalado
```

```python
# ==== 4. CLUSTERING CON K-MEANS ==== #
k_values = range(2, 11)  # Probaremos de 2 a 10 grupos
inertias = []
```

```python
# Iteramos sobre cada valor de k y guardamos la inercia
for k in k_values:
    modelo = KMeans(n_clusters=k, random_state=42)
    modelo.fit(df_scaled.drop('Attrition_rate', axis=1))  # "datos" serán tus columnas escaladas (sin Attrition_rate)
    inertias.append(modelo.inertia_)
```

```python
df_model.select_dtypes(exclude=[np.number])  # Debería estar vacío
```

```python
# ==== 5. ELBOW METHOD ==== #
plt.plot(range(2, 11), inertias, marker='o')
plt.title('Método del codo para elegir k')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.grid(True)
plt.show()
```
